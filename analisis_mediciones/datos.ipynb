Visualización de datos obtenidos por sensores.

A continuación encontrara el codigo utilizado para crear una representación grafica de los datos obtenidos del dispositivo Iot Agriculture measurement, el cual estuvo recopilando datos desde el 22/05/2021 a las 21:20 hasta el 25/05/2021 a las 23:05.

```python
#se importan las librerias necesarias
import pandas as pd
import plotly.express as px
import dash
import datetime
import plotly.graph_objects as go
from ipywidgets import widgets
import numpy as np
```

```python
#dataframe
df = pd.read_csv("data.csv") #se lee el archivo obtenido con los datos recolectados por los sensores. 
df.head(45000) #Imprime una visualización inicial de los datos.
```

```python
print("Procesando la información...")

grouped_data = df.groupby(['Variables', 'M', 'Valores'])
grouped_data.describe() #se obtiene un analisis estadistico basico de las columnas del csv.
```

```python
df = df.drop(df.columns[[0]], axis=1)
variables = ['lux', 'temp', 'awita', 'pressure', 'hum']

dias = widgets.IntSlider(
    value=0.1,
    min=22,
    max=25,
    step=1.0,
    description='Dias:',
    continuous_update=False
)


use_date = widgets.Checkbox(
    description='Fecha: ',
    value=True,
)

container = widgets.HBox(children=[use_date, dias])

origin = widgets.Dropdown(
    options= variables,
    value ='lux',
    description='Variables',
)


# Assign an empty figure widget with two traces
trace2 = go.Histogram(x=df['Valores'],  opacity=0.75, name='Valores')
g = go.FigureWidget(data=[trace2],
                    layout=go.Layout(
                        title=dict(
                            text='Iot Agriculture measurement'
                        ),
                        barmode='overlay'
                    ))
```

```python
dias2 = []
i = 22
while(i <= 25):
    dias2.append(i)
    i += 1
    
def validate():
    if origin.value in df['Variables'].unique():
        return True
    else:
        return False


def response(change):
    if validate():
        if use_date.value:
            filter_list = [i and j for i, j in
                           zip(dias2 == dias.value, 
                               df['Variables'] == origin.value)]
            temp_df = df[filter_list]

        else:
            filter_list = [i and j for i, j in
                           zip(df['Variables'] == 'lux', df['Variables'] == origin.value)]
            temp_df = df[filter_list]
        x1 = temp_df['Valores']
           
        with g.batch_update():
            g.data[0].x = x1
            
            g.layout.barmode = 'overlay'
            g.layout.xaxis.title = 'Dias y horas'
            g.layout.yaxis.title = 'Valores'
```

```python
origin.observe(response, names="value")
dias.observe(response, names="value")
use_date.observe(response, names="value")
```

```python
container2 = widgets.HBox([origin])
widgets.VBox([container, container2,
              g])
```
